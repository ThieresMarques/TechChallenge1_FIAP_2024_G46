```python
# Calculando o ano máximo
ano_maximo = amostra_espumante['Ano'].max()

# Criando um DataFrame para armazenar os próximos 3 anos para cada país
anos_projetados = pd.DataFrame({
  'País': [],
  'Ano': []
})

# Para cada país, adicionamos os próximos 3 anos ao DataFrame de anos projetados
for país in amostra_espumante['País'].unique():
  anos_projetados_país = pd.DataFrame({
      'País': [país] * 5,  # Repetindo o país para os próximos 3 anos
      'Ano': [ano_maximo + i for i in range(1, 6)]
  })
  anos_projetados = pd.concat([anos_projetados, anos_projetados_país], ignore_index=True)

# Criando um DataFrame vazio para armazenar os dados projetados
dados_projetados_espumantes = pd.DataFrame(columns=['País', 'Ano', 'Quantidade', 'Valor'])

# Para cada país, calculamos a projeção para 2024, 2025 e 2026
for país in amostra_espumante['País'].unique():
  # Últimos valores disponíveis de quantidade e valor para o país
  ultimos_dados_origem = amostra_espumante.query('@país in País').tail(5)

  # Projeção para 2024
  quantidade_2024 = ultimos_dados_origem['Quantidade'].mean()
  valor_2024 = ultimos_dados_origem['Valor'].mean()
  dados_projetados_espumantes = pd.concat([dados_projetados_espumantes, pd.DataFrame({'País': [país], 'Ano': [2024], 'Quantidade': [quantidade_2024], 'Valor': [valor_2024]})], ignore_index=True)

# Projeção para 2025
  ultimos_dados_2025 = pd.concat([ultimos_dados_origem.tail(4),dados_projetados_espumantes.query('@país in País')])
  quantidade_2025 = ultimos_dados_2025['Quantidade'].mean()
  valor_2025 = ultimos_dados_2025['Valor'].mean()
  dados_projetados_espumantes = pd.concat([dados_projetados_espumantes, pd.DataFrame({'País': [país], 'Ano': [2025], 'Quantidade': [quantidade_2025], 'Valor': [valor_2025]})], ignore_index=True)

  # Projeção para 2026
  ultimos_dados_2026 = pd.concat([ultimos_dados_origem.tail(3),dados_projetados_espumantes.query('@país in País')])
  quantidade_2026 = ultimos_dados_2026['Quantidade'].mean()
  valor_2026 = ultimos_dados_2026['Valor'].mean()
  dados_projetados_espumantes = pd.concat([dados_projetados_espumantes, pd.DataFrame({'País': [país], 'Ano': [2026], 'Quantidade': [quantidade_2026], 'Valor': [valor_2026]})], ignore_index=True)

  # Projeção para 2027
  ultimos_dados_2027 = pd.concat([ultimos_dados_origem.tail(2),dados_projetados_espumantes.query('@país in País')])
  quantidade_2027 = ultimos_dados_2027['Quantidade'].mean()
  valor_2027 = ultimos_dados_2027['Valor'].mean()
  dados_projetados_espumantes = pd.concat([dados_projetados_espumantes, pd.DataFrame({'País': [país], 'Ano': [2027], 'Quantidade': [quantidade_2027], 'Valor': [valor_2027]})], ignore_index=True)

  # Projeção para 2028
  ultimos_dados_2028 = pd.concat([ultimos_dados_origem.tail(1),dados_projetados_espumantes.query('@país in País')])
  quantidade_2028 = ultimos_dados_2028['Quantidade'].mean()
  valor_2028 = ultimos_dados_2028['Valor'].mean()
  dados_projetados_espumantes = pd.concat([dados_projetados_espumantes, pd.DataFrame({'País': [país], 'Ano': [2028], 'Quantidade': [quantidade_2028], 'Valor': [valor_2028]})], ignore_index=True)

# Agora 'dados_projetados' contém os dados projetados para os próximos 3 anos para cada país
dados_projetados_espumantes['Quantidade'] = dados_projetados_espumantes['Quantidade'].astype(int)
dados_projetados_espumantes['Valor'] = dados_projetados_espumantes['Valor'].astype(int)


```

```python
amostra_espumante['Real_Projecao'] = 'Real'
dados_projetados_espumantes['Real_Projecao'] = 'Projecao'
espumantes = pd.concat([amostra_espumante, dados_projetados_espumantes]).reset_index().sort_values(by=['País','Ano'])
espumantes.drop('index', axis=1, inplace = True)
espumantes['Tipo_dado'] = 'Espumantes'
espumantes.query('País == "Índia"')
```
